```python
import msprot
import numpy as np
u = msprot.Universe('workdir/step2_mutate.dms')
```

```python
u.atoms
```

```python
three2one = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
             'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N',
             'GLY': 'G', 'HIS': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W',
             'ALA': 'A', 'VAL': 'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M',
             'HIP': 'H', 'HSD': 'H', 'HSE': 'H', 'HID': 'H', 'HIE': 'H'}
```

```python
### CO
saved = []
for index, atom in u.atoms.iterrows():
    resid   = atom.resid
    chain   = atom.chain
    resname = atom.resname
    record  = chain + ':' + str(resid)

    if resname not in three2one.keys():
        continue

    bA1 = u.atoms.chain == chain
    bA2 = u.atoms.resid == resid + 1
    bA3 = u.atoms.resid == resid - 1
    bA4 = u.atoms['name'].isin(['C', 'O', 'N', 'HN'])
    
    

    if len(u.atoms[bA1 & bA2]) > 0:
        saved.append(record)
        continue
        
    bA2 = u.atoms.resid == resid - 1
    if len(u.atoms[bA1 & bA2]) > 0:
        saved.append(record)
        continue
        
    bA3 = u.atoms.resid == resid
    bA4 = u.atoms['name'].isin(['C', 'O', 'N', 'HN'])
    
    u.atoms.loc[bA1 & bA3 & bA4, 'bfactor'] = 0.0
    saved.append(record)
```

```python
np.sum(u.atoms.bfactor < 0.5)
```
